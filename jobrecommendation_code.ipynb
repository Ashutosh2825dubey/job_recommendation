Importing Library Files

```python
import pandas as pd
import numpy as np
import nltk
from collections import Counter
import string
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data=pd.read_csv("/content/dataanalystskillrefined1.csv")
```

```python
data.head()
```

Preprocessing


```python

data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data=data.drop(['Unnamed: 0'],axis=1)
data=data.drop(['Unnamed: 0.1'],axis=1)
```

```python
data['transformed_data'] = data['transformed_data'].apply(lambda x:''.join(x))

# Remove leading and trailing spaces and brackets
data['transformed_data'] = data['transformed_data'].str.strip('[]')

# Remove extra commas and spaces
data['transformed_data'] = data['transformed_data'].str.replace(r'\s*,\s*', ',')

# Remove double quotes
data['transformed_data'] = data['transformed_data'].str.replace("'", "")

# Check the data type of the column after conversion
print(data['transformed_data'].dtype)

# Display the DataFrame
print(data)
```

```python

 concatenated_data= data.sample(frac=1).reset_index(drop=True)


print(concatenated_data)
```

```python


def remove_punctuation_and_spaces(text):
    # Remove punctuation
    text_no_punct = text.translate(str.maketrans('', '', string.punctuation))

    # Replace multiple spaces with a single space
    text_cleaned = ' '.join(text_no_punct.split())

    return text_cleaned

# Example usage:
# Assuming cloned_df['transformed_data'] contains your data with punctuations
concatenated_data['transformed_data'] = concatenated_data['transformed_data'].apply(remove_punctuation_and_spaces)
```

```python
concatenated_data
```

```python
cloned_df = concatenated_data[['transformed_data']].copy()
cloned_df
```

```python
user_input = "python excel powerpoint java html css"

# Convert user input to DataFrame and append
new_row = pd.DataFrame({'transformed_data': [user_input]})
cloned_df = pd.concat([cloned_df, new_row], ignore_index=True)
cloned_df
```

Vectorization


```python
cv = TfidfVectorizer(max_features=100,)
vector = cv.fit_transform(cloned_df['transformed_data']).toarray()


```

```python
vector
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
last_row = vector[-1].reshape(1, -1)
last_row
```

```python

similarities = cosine_similarity(last_row,vector[:-1])
```

```python
distances1 = sorted(list(enumerate(similarities[0])),reverse=True,key = lambda x: x[1])[0:655]
```

```python
distances1
```

```python
roles=[]
for i in distances1:
  roles.append(concatenated_data.iloc[i[0]].designation)

```
